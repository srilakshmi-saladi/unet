```python
from pypdf import PdfReader

pages = []
for fname in pdf_files:
    reader = PdfReader(fname)
    for i, page in enumerate(reader.pages):
        txt = (page.extract_text() or "").strip()
        if txt:
            pages.append({"text": txt, "source": fname, "page": i+1})

print("Pages with text:", len(pages))
if pages:
    print("Example:", pages[0]["source"], "p.", pages[0]["page"])
```

```python
def chunk_text(text, chunk_size=1000, overlap=200):
    chunks = []
    start = 0
    while start < len(text):
        end = min(len(text), start + chunk_size)
        chunks.append(text[start:end])
        start = max(end - overlap, start + 1)
    return chunks

chunks = []
for p in pages:
    for c in chunk_text(p["text"], chunk_size=1000, overlap=200):
        chunks.append({"text": c, "source": p["source"], "page": p["page"]})

print("Chunks:", len(chunks))
```

```python
import faiss

texts = [c["text"] for c in chunks]
emb = embedder.encode(
    texts,
    convert_to_numpy=True,
    normalize_embeddings=True,
    show_progress_bar=True
).astype("float32")

index = faiss.IndexFlatIP(emb.shape[1])  # cosine via normalized vectors
index.add(emb)

print("Indexed vectors:", index.ntotal)
```

```python
from transformers import pipeline
gen = pipeline("text2text-generation", model="google/flan-t5-large", max_new_tokens=256)

def rag_answer(question, k=5):
    q_emb = embedder.encode([question], convert_to_numpy=True, normalize_embeddings=True).astype("float32")
    scores, ids = index.search(q_emb, k)

    retrieved = [chunks[i] for i in ids[0]]
    context = "\n\n---\n\n".join(
        [f'[{r["source"]} p.{r["page"]}]\n{r["text"]}' for r in retrieved]
    )

    prompt = f"""Answer using ONLY the context.
If the answer isn't in the context, say: "I don't know from the provided PDFs."
Cite sources like (file p.#).

Question: {question}

Context:
{context}

Answer:"""

    answer = gen(prompt)[0]["generated_text"]
    sources = [(r["source"], r["page"]) for r in retrieved]
    return answer, sources

rag_answer("What is this document about?")
```

```python
import re

def extract_contact_info(all_text):
    # Email
    email = re.findall(r"[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}", all_text)
    email = email[0] if email else None

    # Phone (fairly flexible)
    phone = re.findall(r"(\+?\d[\d\-\s().]{8,}\d)", all_text)
    phone = phone[0] if phone else None

    # Name heuristic: first non-empty line with letters (often top of resume)
    lines = [ln.strip() for ln in all_text.splitlines() if ln.strip()]
    name = None
    for ln in lines[:10]:
        if len(ln) <= 50 and re.search(r"[A-Za-z]", ln) and not re.search(r"@", ln):
            name = ln
            break

    return {"name": name, "email": email, "phone": phone}

# Combine full extracted text
full_text = "\n".join([p["text"] for p in pages])
extract_contact_info(full_text)
```

```python
import pdfplumber

pages = []

for fname in pdf_files:
    with pdfplumber.open(fname) as pdf:
        for i, page in enumerate(pdf.pages):
            txt = (page.extract_text() or "").strip()
            if txt:
                pages.append({
                    "text": txt,
                    "source": fname,
                    "page": i + 1
                })

print("Pages extracted:", len(pages))
print("\n--- SAMPLE TEXT ---\n")
print(pages[0]["text"][:800])
```

```python
def chunk_text(text, chunk_size=800, overlap=150):
    out = []
    start = 0
    while start < len(text):
        end = min(len(text), start + chunk_size)
        out.append(text[start:end])
        start = max(end - overlap, start + 1)
    return out

chunks = []
for p in pages:
    for c in chunk_text(p["text"]):
        chunks.append({"text": c, "source": p["source"], "page": p["page"]})

len(chunks)
```

```python
import faiss
from sentence_transformers import SentenceTransformer

embedder = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")
emb = embedder.encode([c["text"] for c in chunks],
                      convert_to_numpy=True,
                      normalize_embeddings=True).astype("float32")

index = faiss.IndexFlatIP(emb.shape[1])
index.add(emb)
```

```python
def retrieve(question, k=6, fetch_k=30):
    q = embedder.encode([question], convert_to_numpy=True, normalize_embeddings=True).astype("float32")
    scores, ids = index.search(q, fetch_k)

    seen = set()
    results = []
    for i in ids[0]:
        r = chunks[i]
        key = (r["source"], r["page"])
        if key in seen:
            continue
        seen.add(key)
        results.append(r)
        if len(results) >= k:
            break
    return results
```

```python
from transformers import pipeline
gen = pipeline("text2text-generation", model="google/flan-t5-large", max_new_tokens=256)

def ask(question):
    retrieved = retrieve(question, k=6)

    context = "\n\n---\n\n".join(
        [f'[{r["source"]} p.{r["page"]}]\n{r["text"]}' for r in retrieved]
    )

    prompt = f"""You are answering questions about the provided PDF(s).
Use ONLY the context. Be direct.
If the answer is present, answer it and cite (file p.#).
If missing, say: "Not found in the PDF."

Question: {question}

Context:
{context}

Answer:"""

    answer = gen(prompt)[0]["generated_text"]
    sources = [(r["source"], r["page"]) for r in retrieved]
    return answer, sources
```

```python
ask("What is the name on the resume?")
ask("What is the phone number?")
ask("List the work experience.")
```
